```python
'''
#根据事件记录定义和事件记录生成：
is the trigger/argument mention 'dropped' a subsequence of the passage?
is the trigger 'dropped' used to initiate occurrence 'TRANSPORT'?
Is '' a 'TRANSPORT' argument describing the event triggered by 'dropped'? # 记录中没有argument的值就不生成，下同
is the passage containing information that could serve as an argument 'Agent' for event triggered by 'dropped' that should not appear?
is the passage containing information that could serve as an argument 'Artifact' for event triggered by 'dropped' that should not appear?
is the passage containing information that could serve as an argument 'Vehicle' for event triggered by 'dropped' that should not appear?
is the passage containing information that could serve as an argument 'Origin' for event triggered by 'dropped' that should not appear?
is the passage containing information that could serve as an argument 'Destination' for event triggered by 'dropped' that should not appear?

Check the list upon step by step, then return you answer with this dict in json(Vulue allowed: Yes/No):
#然后生成对应的json字典（值填充为"Yes"）：
    {
    "mention_exist":{
            "dropped": "Yes"
            },
    "trigger_initiate_occurrence": "Yes", 
    "argument_describing_event":{}, 
    "argument_should_not_appear":{
            "Agent":"No",
    		"Artifact": "No",
    		"Vehicle": "Yes",
    		"Origin": "No",
    		"Destination": "No"
            }
    }
#根据以上字典生成以下问题：
missing mention 'xxx', add 'xxx' to the passage, and tag it as <Agent>xxx</Agent> # e.g., missing mention 'xxx', add 'xxx' to the passage, and tag it with <Agent></Agent>
event 'TRANSPORT' is not triggered by '', rewirte passage to make sure trigger '' initiate occurrence 'TRANSPORT'
The passage contains a hallucinated argument 'Agent' incorrectly, remove 'Artifact' information for event triggered by 'dropped'.
The passage contains a hallucinated argument 'Artifact' incorrectly, remove 'Artifact' information for event triggered by 'dropped'.
The passage contains a hallucinated argument 'Origin' incorrectly, remove 'Artifact' information for event triggered by 'dropped'.
The passage contains a hallucinated argument 'Destination' incorrectly, remove 'Artifact' information for event triggered by 'dropped'.

'''
```

```python
import json
import os
# Given event definition and event record
fp = './meta_data/arg_roles/'

arg_role_definition = {}
for fn in os.listdir(fp):
    if not fn.endswith('.txt'):
        continue
    with open(fp + fn, 'r', encoding = 'utf-8') as f:
        s = f.read()
    j = json.loads(s)
    arg_role_definition.update(j)
```

```python
def generate_result_json(event_data, event_definitions):
    results = {}
    
    event_type = event_data['event']
    trigger = event_data['trigger']
    arguments = event_data['argument']
    
    # Get roles from event definition
    definition_roles = event_definitions.get(event_type, {})
    simplified_roles = [role.split('-')[0] for role in definition_roles.keys()]
    
    # 1. Build mention_exist
    mention_exist = {trigger: "Yes"}
    for role, mentions in arguments.items():
        for mention in mentions:
            mention_exist[mention + '_of_' + role] = "Yes"
    
    # 2. trigger_initiate_occurrence
    trigger_initiate_occurrence = "Yes"
    
    # 3. argument_describing_event
    argument_describing_event = {}
    argument_describing_event[trigger] = 'Yes'
    for role, mentions in arguments.items():
        for mention in mentions:
            argument_describing_event[mention + '_of_' + role] = 'Yes' 
    
    # 4. argument_should_not_appear
    argument_should_not_appear = {}
    for role in simplified_roles:
        if role not in arguments.keys():
            argument_should_not_appear[role] = "Yes"
    
    # Build final result structure
    results = {
        "mention_exist": mention_exist,
        "trigger_initiate_occurrence": trigger_initiate_occurrence,
        "argument_describing_event": argument_describing_event,
        "argument_should_not_appear": argument_should_not_appear
    }
    
    return results
```

```python
def generate_questions(event_data, arg_role_definition):
    """Generate validation questions for an event record"""
    questions = []
    event_type = event_data['event']
    trigger = event_data['trigger']
    arguments = event_data['argument']
    
    # Extract and simplify role names from definition
    roles = list(arg_role_definition[event_type].keys())
    simple_roles = [role.split('-')[0] for role in roles]
    
    # Base questions about trigger
    questions.extend([
        f"is the trigger mention '{trigger}' a subsequence of the passage?",
        f"is the trigger '{trigger}' used to initiate occurrence '{event_type}'?"
    ])
    for role in arguments:
        for arg in arguments[role]:
            questions.append(f"Is '{arg}' a argument of role '{role}' describing the event '{event_type}' triggered by '{trigger}'?",)
    # Role-specific questions
    for role in simple_roles:
        if role not in arguments.keys():
            questions.append(
                f"is the passage containing information that could serve as an argument '{role}' for event triggered by '{trigger}' that should not appear?"
            )
    q = 'for event {} triggered by {}'.format(event_type, trigger)
    q = '\n'.join(questions)
    
    result_json = generate_result_json(event_data, arg_role_definition)
    q += "\nCheck the list upon, then only return your answer with this dict in json(Value allowed: Yes/No):\n"
    q += json.dumps(result_json, indent=4)
    
    return q

# Output the questions
event_data = {'event': 'ARREST-JAIL',
  'trigger': 'went',
  'argument': {'Place': ['Building'],
   'Person': ['Samiel aryan'],
   'Agent': ['Israel']}}
questions =  generate_questions(event_data, arg_role_definition)

# Part 2: Generate JSON response

print(questions)
```

```python
result = {
    "mention_exist": {
        "what": "Yes",
        "Extremist": "Yes"
    },
    "trigger_initiate_occurrence": "Yes",
    "argument_describing_event": {
        "what": "No",
        "Extremist_of_Target": "No"
    },
    "argument_should_not_appear": {
        "Attacker": "Yes",
        "Target": "No",
        "Instrument": "Yes",
        "Place": "Yes"
    }
}

def generate_problem_statements(result, event_data, arg_role_definition):
    problem_statements = []
    event_type = event_data['event']
    trigger = event_data['trigger']
    arguments = event_data['argument']
    
    # Special trigger validation
    if result["trigger_initiate_occurrence"] == "No":
        problem_statements.append(
            f"event '{event_type}' is not properly triggered by '{trigger}', rewrite passage to ensure '{trigger}' correctly initiates occurrence '{event_type}'"
        )
    
    for arg_role, mention_exist in result["mention_exist"].items():
        if mention_exist.upper() == "No".upper() and len(arg_role.split('_of_')) == 2:
            arg, role = arg_role.split('_of_')
            for role_, args in arguments.items():
                if arg in args:
                    role = role_
            problem_statements.append(
                f"missing mention '{arg}' for event '{event_type}' triggered by '{trigger}', add this mention '{arg}' tagged as <{role}>{arg}</{role}> and appropriate content"
            )
    
    for arg_role, describing_event in result["argument_describing_event"].items():
        if arg_role == trigger:
            problem_statements.append(
                f"event '{event_type}' is missing trigger '{trigger}', rewrite passage to ensure '{trigger}' correctly initiates occurrence '{event_type}'"
            )
            continue
        if describing_event.upper() == "No".upper() and len(arg_role.split('_of_')) == 2:
            arg, role = arg_role.split('_of_')
            problem_statements.append(
                f"'{event_type}' argument '{arg}' is not describing the event triggered by '{trigger}', rewrite passage to ensure '{arg}' correctly describes occurrence '{event_type}' triggered by '{trigger}'"
            )
    
    # Check for hallucinated arguments
    for role, should_not_appear in result["argument_should_not_appear"].items():
        if should_not_appear == "Yes":
            problem_statements.append(
                f"The passage contains a hallucinated argument '{role}' incorrectly, remove '{role}' information for event triggered by '{trigger}'."
            )
    return problem_statements

```

```python
generate_problem_statements(result, event_data, arg_role_definition)
```
